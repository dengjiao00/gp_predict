```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = np.load('result.npz')
num_epoches = data['num_epoches']
```

```python
# 绘制训练loss和测试loss的曲线
train_loss_list = data['train_loss_list']
test_loss_list = data['test_loss_list']


plt.plot(range(1, len(train_loss_list) + 1), train_loss_list, label='train loss')
plt.plot(range(1, len(test_loss_list) + 1), test_loss_list, label='test loss')
plt.title('Loss')
plt.xlabel('loss')
plt.ylabel('epoch')
plt.legend()
plt.show()
```

```python
mae_list = data['mae_list']

# 画出MAE图像
plt.plot(range(1, num_epoches + 1), mae_list, label='mae')
plt.title('MAE')
plt.xlabel('mae')
plt.ylabel('epoch')
plt.legend()
plt.show()
```

```python
rmse_list = data['rmse_list']

# 画出RMSE图像
plt.plot(range(1, num_epoches + 1), rmse_list, label='rmse')
plt.title('RMSE')
plt.xlabel('rmse')
plt.ylabel('epoch')
plt.legend()
plt.show()
```

```python
y_hat_list = data['y_hat_list'].reshape(-1).tolist()
data_y_list = data['data_y_list'].reshape(-1).tolist()

x_len = len(data_y_list)

plt.plot(range(1, x_len + 1), data_y_list, label='true value')
plt.plot(range(1, x_len + 1), y_hat_list, label='predict value')
plt.title('true/pred')
plt.legend()
plt.show()
```
